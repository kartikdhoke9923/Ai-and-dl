```python
# downscaling - preserve aspect ratio
# for grayscale -0 ,for color -1, for alpha = -1
img= cv2.imread('catdog.jpg', 1)
plt.imshow(img)
print(img.shape)
```

```python
# downscaling - preserve aspect ratio
# for grayscale -0 ,for color -1, for alpha = -1
img= cv2.imread('catdog.jpg', 1)
scale_percent= 60
width= int(img.shape[1] * scale_percent/100)
height= int(img.shape[0] * scale_percent/100)
dim=(width, height)
resized= cv2.resize(img, dim, interpolation=cv2.INTER_LINEAR)
plt.imshow(img)
print(resized.shape)
```

```python
# rotating image
img=cv2.imread('catdog.jpg',1)
h,w = img.shape[:2]
print(w, h)
center= (w//2, h//2)
matrix = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated= cv2.warpAffine(img, matrix, (w,h))
plt.imshow(img)
plt.imshow(rotated)
```

```python
# edge detection perform only on grey scale by reducing complexity
```

```python
img=cv2.imread('catdog.jpg', 1)
greyscale_img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(greyscale_img)
edges= cv2.Canny(greyscale_img, threshold1=80, threshold2=250)
plt.imshow(edges)
```
