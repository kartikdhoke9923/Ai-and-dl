````python
docs = ['go india',
        'india india',
        'hip hip hurray',
        'jeetega bhai jeetega india jeetega',
        'bharat mata ki jai',
        'kohli kohli',
        'sachin sachin',
        'dhoni dhoni',
        'modi ji ki jai',
        'inquilab zindabad']
````

````python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()    # Splits text
# Cleans text (optional)
# Converts words → numbers (critical)
# Handles unknown words
# Pads sequences
````

````python
tokenizer.fit_on_texts(docs)
````

````python
len(tokenizer.word_index)
````

````python
from typing import Sequence
sequence=tokenizer.texts_to_sequences(docs)
sequence
````

````python
from keras.utils import pad_sequences
sequences = pad_sequences(sequence, padding='post')
````

````python
sequences
````

### pad_sequences Explanation

pad_sequences(sequence, padding='post') ensures all sequences have the same length.

````python
from keras import Sequential
from keras.layers import Embedding
from tensorflow.keras import layers

model=Sequential()
model.add(Embedding(input_dim=18, output_dim=2))  #input_dim=18     because of  length is 17
model.summary()
````

````python
model.compile('adam', 'accuracy')
````

````python
pred=model.predict(sequences)
````
