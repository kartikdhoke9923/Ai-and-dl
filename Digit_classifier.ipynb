```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import tensorflow as tf
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()    # dataset by yaan li yun
```

```python
X_train.shape # 60000 image each image = 28 × 28 pixels
```

```python
X_train[0]
```

```python

plt.matshow(X_train[0])
```

```python
X_train=X_train/255
```

```python
X_test=X_test/255
```

```python
X_train[0]
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)
```

```python
model=keras.Sequential(
    [keras.layers.Dense(15, input_shape=(784,),activation='sigmoid'),
     keras.layers.Dense(15, input_shape=(784,),activation='softmax')]
)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train_flattened,y_train, epochs=10)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```

```python
plt.matshow(X_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels= [np.argmax(i) for i in y_predicted]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predicted_labels)
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions= y_predicted_labels)
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predictred')
plt.ylabel('Truth')
```
