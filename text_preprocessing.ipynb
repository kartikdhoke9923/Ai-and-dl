## text normalization (stemming)
- 1 breaking the word to its stem
eg study , studying , studied

```python
# using nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem import PorterStemmer

word_token = word_tokenize(text)
stopwords=set(stopwords.words('english'))

punc=list(punctuation)

filtered_tokens=[]
for w in word_token:
  if w not in stopwords:
    if w not in punc:
      filtered_tokens.append(w)

print('orignal tookens', word_token)
print('filtered tokens', filtered_tokens)

# stemming
stemmed_token=[]
ps=PorterStemmer()
for w in filtered_tokens:
  root_words= ps.stem(w)
  stemmed_token.append(root_words)

print('orignal tokens', word_token)
print('filtered tokens', filtered_tokens)
print('stemmed token', stemmed_token)
```

## lemmatization

```python
# example work
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
words=['study', 'geese','better','sleeping','mice']
lemmas_verb= [wl.lemmatize(word, pos="v") for word in words]
lemmas_noun= [wl.lemmatize(word, pos="n") for word in words]    # here n = noun , v= vverb, a = adjective
lemmas_adjective= [wl.lemmatize(word, pos="a") for word in words]

print(lemmas_verb)
print(lemmas_noun)
print(lemmas_adjective)
```

```python
from gensim.models import Word2Vec

corpus = ['machine learning is powerful',
          'deep learning is a part of machine learning',
          'machine learning is fun',
          'deep learning includes neural networks',
          'neural networks are the base of the transformers']
sentence=[sentence.split() for sentence in corpus]
```

```python
w2v_model= Word2Vec(sentence, vector_size=300, window=2, min_count=1, workers=4)
```

```python
import numpy as np

def vectorizer(sentence):
  words= sentence.split()
  words_vec= [w2v_model.wv[word] for word in words if word in w2v_model.wv]
  if len(words_vec)==0:
    return np.zeros(300)
  words_vec=np.array(words_vec)
  return words_vec.mean(axis=0)
```

```python
corpus_embeded= np.array([vectorizer(sentence) for sentence in corpus])
```
