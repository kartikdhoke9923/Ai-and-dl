```python
import cv2
import pytesseract

import matplotlib.pyplot as plt
```

```python
images=cv2.imread('text_im.png',1)
if images is None:
  print('Image not found')
else:
  print('loaded succesfully')
```

```python
print(images.shape) #its already gray scale
```

```python
gray=cv2.cvtColor(images, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
thresh= cv2.threshold(gray, 150, 225, cv2.THRESH_BINARY_INV)[1]
plt.imshow(thresh)
```

```python
resized = cv2.resize(thresh, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(resized)
```

```python
# niw reducung noice
denoised=cv2.fastNlMeansDenoising(resized,None, 30, 7, 21)
plt.imshow(denoised)
```

```python
text=pytesseract.image_to_string(denoised, lang='eng')
```
